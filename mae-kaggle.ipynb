```python
#@title

import os, sys
def get_run_type() -> str:
    if os.environ.get('KAGGLE_KERNEL_RUN_TYPE'):
        return 'kaggle'
    elif 'google.colab' in sys.modules:
        return 'colab'
    else:
        raise Exception('Unsupported run type')

def clone_or_pull_repo():
    if os.path.isdir(repo_dir_path):
        %cd $repo_dir_path
        !git pull
    else:
        %cd $root_dir
        !git clone https://github.com/thinh-re/mae mae
        %cd $repo_dir_path

run_type = get_run_type()
        
if run_type == 'kaggle':
    root_dir = '/kaggle/working'
elif run_type == 'colab':
    root_dir = '/content'
else:
    raise Exception('Unsupported run type')

repo_dir_path = os.path.join(root_dir, 'mae')

if run_type == 'colab':
    if not os.path.isdir('/content/drive'):
        from google.colab import drive
        drive.mount('/content/drive')
```
